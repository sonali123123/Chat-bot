```python

import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
webpage = requests.get('https://www.encorebostonharbor.com/amenities/the-spa').text
soup = BeautifulSoup(webpage,'lxml')
```

```python
print(soup.prettify())
```

```python
questions = []

for i in soup.find_all('div',class_='questions-title'):
    questions.append(i.find('span', class_='question-title-clickable').text.strip())

questions
```

```python
answers= []
for i in soup.find_all('p'):
    answers.append(i.text.strip())

answers = [i for i in answers if i != ''][4:-8]
answers
```

```python
len(questions), len(answers)
```

```python
faq_dict = dict(zip(questions, answers))
faq_dict
```

```python
Spa_df = pd.DataFrame(list(faq_dict.items()), columns=["Question", "Answer"])
Spa_df
```

```python
webpage = requests.get("https://www.encorebostonharbor.com/amenities/the-salon").text
soup = BeautifulSoup(webpage,'lxml')
```

```python
print(soup.prettify())
```

```python
questions = []

for i in soup.find_all('div',class_='questions-title'):
    questions.append(i.find('span', class_='question-title-clickable').text.strip())
questions
```

```python
answers= []
for i in soup.find_all('p'):
    answers.append(i.text.strip())
answers = [i for i in answers if i != ''][6:-8]
answers
```

```python
len(questions), len(answers)
```

```python
faq_dict = dict(zip(questions,answers))
```

```python
Salon_df = pd.DataFrame(list(faq_dict.items()), columns=["Question", "Answer"])
Salon_df
```

```python
import pandas as pd

# Assuming you have Spa_df and Salon_df dataframes

# Merge the dataframes
merged_df = pd.concat([Spa_df, Salon_df], ignore_index=True)

# Remove duplicate rows
merged_df = merged_df.drop_duplicates()

# Print the resulting dataframe
merged_df
```
